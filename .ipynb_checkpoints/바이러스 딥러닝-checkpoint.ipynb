```python
train_target = train_df['class']
test_target = test_df['class']
```

```python
train_dataset = tf.data.Dataset.from_tensor_slices((train_df.values, train_target.values)).shuffle(10).batch(1)
```

```python
test_dataset = tf.data.Dataset.from_tensor_slices((test_df.values, test_target.values)).shuffle(10).batch(1)
```

```python
def get_compiled_model():
  model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(10)
 ])

  model.compile(optimizer='adam',
                loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
                metrics=['accuracy'])
  return model
```

```python
model = get_compiled_model()
model.fit(train_dataset, epochs=10)
```

```python
model.evaluate(test_dataset)
```
